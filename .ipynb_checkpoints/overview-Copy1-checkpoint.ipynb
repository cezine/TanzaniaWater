```python
from pandas_profiling import ProfileReport
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pickle 
```

## Importing data

```python
dfx = pd.read_csv("data/training_set_values.csv", index_col=0)
```

```python
dfy = pd.read_csv("data/training_set_labels.csv", index_col=0)
```

```python
dfx.shape
```

```python
df = pd.concat([dfx, dfy], axis=1)
```

```python
df.columns
```

## Features with NaN values

>funder has 3635 (6.1%) missing values: Who funded the well <br />
>installer has 3655 (6.2%) missing values: Organization that installed the well<br />
>scheme_management has 3877 (6.5%) missing values: Who operates the waterpoint<br />
>permit has 3056 (5.1%) missing values: If the waterpoint is permitted <br />

>dataset is large enough to lose less than 10% of the data

```python
df.isna().sum()
```

## Simlilar and extraneous featrues 

> removing similar features:  <br />
>'water_quality' and 'quality_group'  <br />
>'quantity' and 'quantity_group'  <br />
>'source', 'source_type' and 'source_class'  <br />
>'waterpoint_type' and 'waterpoint_type_group'

```python
df.drop(['source', 'source_class','waterpoint_type_group', 'quantity_group', 'payment_type', 
         'management_group','extraction_type_class','extraction_type_group', 'scheme_name', 'recorded_by', 'public_meeting', 
         'region_code', 'wpt_name', 'date_recorded','quality_group','num_private','funder','installer','subvillage','scheme_management','permit'], axis= 1,inplace= True)
```

```python
df.shape
```

## Plots

```python
ig, ax = plt.subplots(figsize=(15,8))
sns.set_style("whitegrid", {'axes.grid' : False})
a= sns.histplot(x=df['quantity'], data=df, hue ='status_group',multiple="dodge",stat='density')

ax.set_xlabel("The quantity of water", fontsize=14, labelpad=20)
ax.set_ylabel("Percentage of Wells", fontsize=18, labelpad=80, rotation=0)
ax.set_title("The quantity of water VS. Percentage of Wells", fontsize=20, pad=15);
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(title='Well Conditions', loc='upper right', labels=['Needs Repairs', 'Not Functional', 'Functional'])

#plt.yticks(fontsize=14)
for p in a.patches:
    a.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()),
               ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
    
#plt.savefig('quantity_vs_well.png', dpi=100, bbox_inches='tight')
```

```python
ig, ax = plt.subplots(figsize=(20,8))
sns.set_style("whitegrid", {'axes.grid' : False})
a= sns.histplot(x=df['waterpoint_type'], data=df, hue ='status_group',multiple="dodge",stat='density')

ax.set_xlabel("Waterpoint Type", fontsize=18, labelpad=20)
ax.set_ylabel("Percentage of Wells", fontsize=18, labelpad=80, rotation=0)
ax.set_title("Waterpoint Type VS. Percentage of Wells", fontsize=20, pad=15);
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(title='Well Conditions', loc='upper right', labels=['Needs Repairs', 'Not Functional', 'Functional'])

for p in a.patches:
    a.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()),
               ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
    
#plt.savefig('watertype_vs_well.png', dpi=100, bbox_inches='tight')
```

```python
ig, ax = plt.subplots(figsize=(20,8))
sns.set_style("whitegrid", {'axes.grid' : False})
a= sns.histplot(x=df['payment'], data=df, hue ='status_group',multiple="dodge",stat='density')

ax.set_xlabel("Payment Type", fontsize=18, labelpad=20)
ax.set_ylabel("Percentage of Wells", fontsize=18, labelpad=80, rotation=0)
ax.set_title("Payment Type VS. Percentage of Wells", fontsize=20, pad=15);

plt.legend(title='Well Conditions', loc='upper right', labels=['Needs Repairs', 'Not Functional', 'Functional'])

for p in a.patches:
    a.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()),
               ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

#plt.savefig('paymenttype_vs_well.png', dpi=100, bbox_inches='tight')
```

```python
ig, ax = plt.subplots(figsize=(15,8))
sns.set_style("whitegrid", {'axes.grid' : False})
a= sns.histplot(x=df['status_group'], data=df)

ax.set_xlabel("Well Status", fontsize=18, labelpad=20)
ax.set_ylabel("Count", fontsize=18, labelpad=80, rotation=0)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
for p in a.patches:
    a.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()),
               ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
    
#plt.savefig('status_group.png', dpi=100, bbox_inches='tight')
```

```python
ig, ax = plt.subplots(figsize=(15,8))
sns.set_style("whitegrid", {'axes.grid' : False})
a= sns.histplot(x=df['construction_year'], data=df, hue ='status_group',multiple="dodge",stat='count', binwidth=8)

ax.set_xlabel("Construction Year", fontsize=18, labelpad=20)
ax.set_ylabel("Number of Wells", fontsize=18, labelpad=80, rotation=0)
ax.set_title("Construction Year VS. Number of Wells", fontsize=20, pad=15);
ax.set(xlim=(1950, 2020))
ax.set(ylim=(0, 8000))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(title='Well Conditions', loc='upper right', labels=['Needs Repairs', 'Not Functional', 'Functional'])

    
plt.savefig('construction_year_vs_well.png', dpi=100, bbox_inches='tight')
```
